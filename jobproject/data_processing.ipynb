## Import Modules

```python
import pandas as pd
import re
import string
```

## Convert Text File to Dataframe

```python
with open('companies1.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# 1. Insert newlines AFTER Pty Ltd / Pty. Ltd / Limited / Inc
pattern = r'(?i)(pty\.?\s+ltd|pty\.?\s+limited|ltd|limited|inc)\s+(?=[A-Z])'

fixed_text = re.sub(pattern, r'\1\n', text)

# 2. Split & clean
companies = [line.strip() for line in fixed_text.split('\n') if line.strip()]

companies_df = pd.DataFrame(companies, columns=['company_name'])
```

```python
def lowercase(text):
    text = text.lower()
    return text
```

```python
def remove_char(text):
    text = re.sub(r'[^a-z\s&]', '', text)
    remove = ['pty', 'ltd', 'limited']
    for word in remove:
        text = text.replace(word, "")
    return text
```

```python
companies_df['company_lower'] = companies_df['company_name'].apply(lowercase) 
```

```python
companies_df['company_processed'] = companies_df['company_lower'].apply(remove_char)
```
