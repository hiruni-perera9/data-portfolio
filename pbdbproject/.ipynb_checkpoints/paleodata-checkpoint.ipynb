```python
import requests
import pandas as pd
from typing import Optional, Dict, Any

PALEODB_BASE_URL = "https://paleobiodb.org/data1.2"
```

```python
def search_dinosaur(name: str) -> Dict[str, Any]:
    url = f"{PALEODB_BASE_URL}/taxa/list.json"
    params = {
        "name" : name,
        "rel" : "all_children",
        "show" : "attr, app, size, class"
    }
    
    response = requests.get(url, params=params)
    response.raise_for_status()
    return response.json()
```

```python
data = search_dinosaur("Chordata")
print(f"Found {len(data.get('records', []))} records")
```

```python
if data.get('records'):
    first_record = data['records'][0]
    print("\nFirst record:")
    for key, value in first_record.items():
        print(f"  {key}: {value}")
```
